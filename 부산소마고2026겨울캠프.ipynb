```python
# 뭐 만들라했지
# 2진수 입력 받아 -> 10진수로 변환

bin = input("2진수를 입력해주세요(최대 5자리): ")


print("10진수로 변환을 시작합니다...")
if int(bin) >= 111110:
  print("2진수가 6자리 이상입니다!")
else:
  bin2 = str(bin)
  five = int(bin2[0]) * 16
  four = int(bin2[1]) * 8
  three = int(bin2[2]) * 4
  two = int(bin2[3]) * 2
  one = int(bin2[4]) * 1
  result = five + four + three + two + one
  print("변환값은 {}입니다".format(result))
```

```python
# 뭐 만들라했지
# 2진수 입력 받아 -> 8진수로 변환

bin = input("2진수를 입력해주세요(소수 불가, 최대 5자리): ")


print("8진수로 변환을 시작합니다...")
def BinToOct(bin):
  if int(bin) >= 111110:
    print("2진수가 6자리 이상입니다!")
  else:
    result10 = 0
    if len(bin) >= 1:
      one = int(bin[-1]) * 1
    if len(bin) >= 2:
      two = int(bin[-2]) * 2
    if len(bin) >= 3:
      three = int(bin[-3]) * 4
    if len(bin) >= 4:
      four = int(bin[-4]) * 8
    if len(bin) >= 5:
      five = int(bin[-5]) * 16

    Dec = one + two + three + four + five

    Oct = ""
    while Dec > 0:
      Oct = str(Dec % 8) + Oct
      Dec //= 8

    print("8진수 변환값은 {}입니다".format(Oct))

bin2 = str(bin)
BinToOct(bin2)
```
