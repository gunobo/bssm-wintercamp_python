```python
# 어떤 10진수 n을 입력받고 2진수로 변환해서 출력하기

def to_binary(n):
    if n == 0:
        return "0"

    def recursive_binary(n):
        if n == 0:
            return ""

        return recursive_binary(n // 2) + str(n % 2)

    return recursive_binary(n)

n = int(input())
if n == 0:
    print(0)
else:
    print(to_binary(n))
```

```python
# 셀프넘버

def d(n):
    return n + sum(map(int, str(n)))

a, b = map(int, input().split())

generated_nums = set()
for i in range(1, b + 1):
    generated_nums.add(d(i))

self_num_sum = 0
for i in range(a, b + 1):
    if i not in generated_nums:
        self_num_sum += i

print(self_num_sum)
```
