```python
#퀵 정렬 파이썬 코드
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8] # 리스트
print("처음 array 리스트:",array) # 출력


#-------------여기부터 한 함수 ---------------#
def quick_sort(arrary,start, end):
  if start >= end:
    return
  pivot = start    # array 리스트
  left = start + 1 # [pivot][start][]][][][][][][end]
  right = end      #
  while left <= right: # left가 right보다 같거나 작을때,
    # 피벗보다 큰 데이터를 찾을 때까지 반복
    while left <= end and arrary[left] <= array[pivot]: # 왼쪽 파티션에서 피벗보다 큰 데이터 찾기전까지 + 1
      left +=1
    #피벗보다 작은 데이터를 찾을때 까지 반복
    while right > start and array[right] >= array[pivot]: # 오른쪽 파티션에서 피벗보다 작은 데이터 찾기전까지 -1
      right -= 1
    if left > right: # 엇갈렸다면 작은 right -= 1 데이터와 피벗을 교체 # -> left가 right 초과한 경우!
      array[right], array[pivot] = array[pivot], array[right] # --> array[right]와 arrary[pivot]을 교체!
    else: # 엇갈리지 않았다면 작은 데이터와 큰 데이터를 교체 # -> left < right 를 유지한다면
      array[left], array[right] = array[right], array[left] # --> array[left]와 array[right]를 교체!

  # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행
  quick_sort(array, start, right-1) # 왼쪽 파트 진행
  quick_sort(array, right+1, end) # 오른쪽 파트 진행
#-------------여기까지 한 함수 ---------------#

quick_sort(array, 0, len(array)-1) # 함수 실행
print("퀵정렬을 한 array 리스트:",array) # 출력
```
